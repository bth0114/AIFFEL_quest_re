학습 시 InvalidArgumentError가 발생하여 크기를 통일시키는 resize 코드 추가함

```python
# Q. 이미지의 표현이 0과 1 사이로 들어오도록 직접 정규화 코드를 작성해봅시다.
def normalize_and_resize_img(image, label):
    image = tf.image.resize(image, [224, 224])  # 크기 통일
    image = tf.cast(image, tf.float32) / 255.0  # 정규화  """Normalizes images: `uint8` -> `float32`."""
    return image, label
```

```python
def apply_normalize_on_dataset(ds, is_test=False, batch_size=16):
    ds = ds.map(
        normalize_and_resize_img,
        num_parallel_calls=1
    )
    ds = ds.batch(batch_size)
    if not is_test:
        ds = ds.repeat()
        ds = ds.shuffle(200)
    ds = ds.prefetch(tf.data.experimental.AUTOTUNE)
    return ds
```

## ResNet 기본 블록 구성

```python
# function for building ResNet basic Block

def build_basic_block(input_layer,
                    num_cnn=3,
                    channel=64,
                    block_num=2,
                    is_plain=False
                   ):
    # 입력 레이어
    x = input_layer

    # CNN 레이어
    for cnn_num in range(1,num_cnn+1):
        skip = x
        if cnn_num == 1 and block_num != 2:
            strides = (2,2)
        else:
            strides = (1,1)

        # 3 x 3
        x = keras.layers.Conv2D(
            filters=channel,
            kernel_size=(3,3),
            strides=strides,
            kernel_initializer='he_normal',
            padding='same',
            name=f'stage{block_num}_{cnn_num}_conv1'
        )(x)
        x = keras.layers.BatchNormalization(name=f'stage{block_num}_{cnn_num}_bn1')(x)
        x = keras.layers.Activation('relu')(x)

        # 3 x 3
        x = keras.layers.Conv2D(
            filters=channel,
            kernel_size=(3,3),
            kernel_initializer='he_normal',
            padding='same',
            name=f'stage{block_num}_{cnn_num}_conv2'
        )(x)
        x = keras.layers.BatchNormalization(name=f'stage{block_num}_{cnn_num}_bn2')(x)

        # skip connection
        if not is_plain:
            if skip.shape[-1] != x.shape[-1]:
                skip = keras.layers.Conv2D(
                        filters=x.shape[-1],
                        kernel_size=(1, 1),
                        strides=strides,
                        padding='same',
                )(skip)
            #skip = keras.layers.BatchNormalization()(skip)
            x = keras.layers.Add(name=f"stage{block_num}_{cnn_num}_add")([x,skip])

        x = keras.layers.Activation('relu')(x)

    return x
```

```python
# function for building ResNet bottleneck Block

def build_bottleneck_block(input_layer,
                    num_cnn=3,
                    channel=64,
                    block_num=2,
                    is_plain=False
                   ):
    # 입력 레이어
    x = input_layer

    # CNN 레이어
    for cnn_num in range(1, num_cnn+1):
        skip = x
        if cnn_num == 1 and block_num != 2:
            strides = (2,2)
        else:
            strides = (1,1)

        # 1 x 1
        x = keras.layers.Conv2D(
            filters=channel,
            kernel_size=(1,1),
            strides=strides,
            kernel_initializer='he_normal',
            padding='same',
            name=f'stage{block_num}_{cnn_num}_conv1'
        )(x)
        x = keras.layers.BatchNormalization(name=f'stage{block_num}_{cnn_num}_bn1')(x)
        x = keras.layers.Activation('relu')(x)

        # 3 x 3
        x = keras.layers.Conv2D(
            filters=channel,
            kernel_size=(3,3),
            kernel_initializer='he_normal',
            padding='same',
            name=f'stage{block_num}_{cnn_num}_conv2'
        )(x)
        x = keras.layers.BatchNormalization(name=f'stage{block_num}_{cnn_num}_bn2')(x)
        x = keras.layers.Activation('relu')(x)

        # 1 x 1
        x = keras.layers.Conv2D(
            filters=channel*4,
            kernel_size=(1,1),
            kernel_initializer='he_normal',
            padding='same',
            name=f'stage{block_num}_{cnn_num}_conv3'
        )(x)
        x = keras.layers.BatchNormalization(name=f'stage{block_num}_{cnn_num}_bn3')(x)

        # skip connection
        if not is_plain:
            if skip.shape[-1] != x.shape[-1]:
                skip = keras.layers.Conv2D(
                        filters=x.shape[-1],
                        kernel_size=(1, 1),
                        strides=strides,
                        padding='same',
                )(skip)
                skip = keras.layers.BatchNormalization()(skip)
            x = keras.layers.Add(name=f"stage{block_num}_{cnn_num}_add")([x,skip])

        x = keras.layers.Activation('relu')(x)

    return x
```

```python
# ResNet 모델 자체를 생성하는 함수입니다.
def build_resnet(input_shape=(32,32,3),
                num_cnn_list=[3,4,6,3],
                channel_list=[64,128,256,512],
                num_classes=10,
                is_plain=False,
                is_50=False):

    assert len(num_cnn_list) == len(channel_list) #모델을 만들기 전에 config list들이 같은 길이인지 확인합니다.

    input_layer = keras.layers.Input(shape=input_shape)  # input layer를 만들어둡니다.
    output = input_layer

    # conv1
    output = keras.layers.Conv2D(filters=64, kernel_size=(7,7), strides=2, kernel_initializer='he_normal', padding='same', name='stage1_conv1')(output)
    output = keras.layers.BatchNormalization(name='stage1_bn')(output)
    output = keras.layers.Activation('relu')(output)

    # Max Pooling 레이어
    output = keras.layers.MaxPooling2D(pool_size=(2, 2), strides=2, name='stage2_0_maxpooling')(output)

    # config list들의 길이만큼 반복해서 블록을 생성합니다.
    for i, (num_cnn, channel) in enumerate(zip(num_cnn_list, channel_list)):
        if not is_50:
            output = build_basic_block(
                output,
                num_cnn=num_cnn,
                channel=channel,
                block_num=i+2,
                is_plain=is_plain
            )
        else:
            output = build_bottleneck_block(
                output,
                num_cnn=num_cnn,
                channel=channel,
                block_num=i+2,
                is_plain=is_plain
            )

    output = keras.layers.AveragePooling2D(pool_size=(1,1), name='avg_pool')(output)
    output = keras.layers.Flatten(name='flatten')(output)
    output = keras.layers.Dense(num_classes, activation='softmax', name='fc1000')(output)

    model = keras.Model(
        inputs=input_layer,
        outputs=output
    )
    return model
```

## ResNet-34 Complete Model

```python
resnet_34 = build_resnet(input_shape=(32,32,3), is_50=False)

resnet_34.summary()
```

## ResNet-50 Complete Model

```python
resnet_50 = build_resnet(input_shape=(32,32,3), is_50=True)

resnet_50.summary()
```

## Ablation Study

학습 중간에 코랩의 RAM 리소스를 초과하여 런타임이 끊어지는 문제가 발생
- 이미지 로드 10000장으로 제한
- 배치 사이즈 32에서 16으로 줄임

### ResNet-50 vs Plain-50

```python
plain_50 = build_resnet(input_shape=(224,224,3), is_50=True, is_plain=True)

plain_50.summary()
```

```python
resnet_50 = build_resnet(input_shape=(224,224,3), is_50=True)

resnet_50.summary()
```

```python
BATCH_SIZE = 16
EPOCH = 15
```

```python
ds_train = apply_normalize_on_dataset(ds_train, batch_size=BATCH_SIZE)
ds_val = apply_normalize_on_dataset(ds_val, batch_size=BATCH_SIZE)
```

```python
plain_50.compile(
    loss='sparse_categorical_crossentropy',
    optimizer=tf.keras.optimizers.SGD(learning_rate=0.01, clipnorm=1.),
    metrics=['accuracy'],
)

steps_per_epoch = train_count // BATCH_SIZE
validation_steps = (total_count-train_count) // BATCH_SIZE

history_p50 = plain_50.fit(
    ds_train,
    steps_per_epoch=steps_per_epoch,
    validation_steps=validation_steps,
    epochs=EPOCH,
    validation_data=ds_val,
    verbose=1,
    #use_multiprocessing=True,
)
```

```python
resnet_50.compile(
    loss='sparse_categorical_crossentropy',
    optimizer=tf.keras.optimizers.SGD(learning_rate=0.01, clipnorm=1.),
    metrics=['accuracy'],
)

history_r50 = resnet_50.fit(
    ds_train,
    steps_per_epoch=steps_per_epoch,
    validation_steps=validation_steps,
    epochs=EPOCH,
    validation_data=ds_val,
    verbose=1,
    #use_multiprocessing=True,
)
```

```python
import matplotlib.pyplot as plt

plt.plot(history_p50.history['loss'], 'r')
plt.plot(history_r50.history['loss'], 'b')
plt.title('Model training loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['plain_50', 'resnet_50'], loc='upper left')
plt.show()
```

```python
plt.plot(history_p50.history['val_accuracy'], 'r')
plt.plot(history_r50.history['val_accuracy'], 'b')
plt.title('Model validation accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['plain_50', 'resnet_50'], loc='upper left')
plt.show()
```

회고
- 노드 3번 ResNet-50 모델 구조 예시에는 average pooling이랑 fc1000이 없는데 논문의 Table 1에는 있어서 논문대로 했다. 이 2개가 없어도 큰 성능 차이가 없는 것인지도 ablation study 해보면 좋을 것 같다.
- skip connection 유무의 차이라서 ResNet 모델의 성능이 더 좋더라도 loss는 떨어질 것이라 생각했는데 Plain 모델의 loss가 15 epoch까지도 거의 떨어지지 않는 이유가 궁금하다. 학습을 50 epochs까지 시켜도 비슷한 경향을 띄는 그래프가 그려질 지도 궁금하다.
- Plain 모델의 loss가 학습 초기에 ResNet에 비해 낮은 이유도 찾아봐야 할 것 같다.
- 주말에 LMS Cloud GPU 연결도 안 되고, 코랩은 DLThon 때 유료 결제까지 해서 다 써서 정말 쥐어짜내서 학습시키느라 분석을 좀 더 못해본 것이 아쉽다. 앞으로 계속 GPU 써야하는 프로젝트가 많을텐데 걱정이다.
